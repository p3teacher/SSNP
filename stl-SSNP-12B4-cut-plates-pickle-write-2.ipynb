```python
# stl-SSNP-12B4-cut-plates-pickle-write-2.py
# 2021-12-28
#
# pickle 是 Python 的內建模組
# read 349x207x2_XZ.dat file  <-- derived from 349x196x3_XYZ.dat (X,Z) elements
#
import os
path = os.getcwd()
print('當前工作目錄 ==>', path)
#
from shapely.geometry  import LineString
from operator import itemgetter
#
import pprint
import pickle
#
path_data    = './data/'
#
def do_A(row,point2D_list) :
    #
    check_sw = "F" 
    for j, xy in enumerate(row) :
        if check_sw == "T" :
            point2D_list.append(row[j])              
        elif row[j+1][1] > 0.0 :
            check_sw = "T"
            point2D_list.append(row[j])   
        # end if
    # end for    
# end def
#
def do_B(row,point2D_list) :
    #
    point2D_list.append(row[-7])
    point2D_list.append(row[-6])
    point2D_list.append(row[-5])
    point2D_list.append(row[-4])
    point2D_list.append(row[-3])
    point2D_list.append(row[-2])
    point2D_list.append(row[-1])
    #
# end def
#  
## ================================================================
#
with open(path_data + '349x207x2_XZ.dat', 'rb') as in_file :
    xz_rows = pickle.load(in_file)
# end with    
#
# trimed xy_rows
xy_rows = []
for i, row in enumerate(xz_rows) :
    y_max = max(row,key=itemgetter(1))[1]
    point2D_list = []
    [point2D_list.append(xy) for xy in row[:6]]
    if y_max > 0.0 :
        do_A(row,point2D_list)
        xy_rows.append(point2D_list)
    else :
        do_B(row,point2D_list)
        xy_rows.append(point2D_list)
    # end if
    #
# end for
#
print('Laser cut boundary has', len(xy_rows), 'rows')
print('         each rows has varying columns')
print ('\n')
#
pprint.pprint(xy_rows[55])
# xy_rows --> shape=(349,varying,2)
with open(path_data + '349-v1-2_XY.dat', 'wb') as out_file :
    pickle.dump(xy_rows, out_file)
# end with    
#
print ('Done')
```
