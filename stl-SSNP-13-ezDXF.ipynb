```python
# stl-SSNP-13-ezDXF.py
# 2021-12-12
#
# pickle 是 Python 的內建模組
# read 349x196x3_XYZ.dat file
#
import os
path = os.getcwd()
print('當前工作目錄 ==>', path)
#
import pickle
import ezdxf
#
path_data    = './data/'
path_graphic = './graphic/'
#
"""
# AC1027 ---> AutoCAD R2013
# line weight ---> 0.13mm = 13
# minimum line weight 13 
# maximum line width 200
#
# color  0 --> Black
# color  1 --> red
# color  3 --> green
# color  5 --> Dark blue
# color  6 --> Magenta
"""
#
## ================================================================
#
with open(path_data + '349x196x3_XYZ.dat', 'rb') as in_file :
    dem_rows = pickle.load(in_file)
# end with    
#
# up layer verteices
A_up = [38.0, 90.0, 0.0] # 0
B_up = [-1.0, 90.0, 0.0] # 1
C_up = [-1.0, 20.4, 0.0] # 2
D_up = [38.0, 20.4, 0.0] # 3
#
# down layer verteices
a_dn = [38.0, 90.0,-2.0] # 4
b_dn = [-1.0, 90.0,-2.0] # 5
c_dn = [-1.0, 20.4,-2.0] # 6
d_dn = [38.0, 20.4,-2.0] # 7
#
left_dn_xy  = (-1.0,-2.0)
left_up_xy  = (-1.0, 0.0)
# ... here is the dem boundary line
right_up_xy = (38.0, 0.0)
right_dn_xy = (38.0,-2.0)
#
left_open_cut_xy  = [(10.15, -2) , (10.15, -1.2), ( 9.85, -1.2), ( 9.85, -2)]
right_open_cut_xy = [(28.15, -2) , (28.15, -1.2), (27.85, -1.2), (27.85, -2)]
#
rows = []
for i in range(0,349,1) :
    XYZ_tuple_row = [(XYZ[0], round(XYZ[2],3)) for XYZ in dem_rows[i]]
    Laser_Cut_Boundary = left_open_cut_xy + [left_dn_xy,left_up_xy] + XYZ_tuple_row + [right_up_xy, right_dn_xy] + right_open_cut_xy + [(10.1, -2)]
    rows.append(Laser_Cut_Boundary)
# end for
#
print('Laser cut boundary has', len(rows),    'rows')
print('         each rows has', len(rows[0]), 'columns')
print ('\n')
#
i = 50
data_rows = rows[i:i+10]
for row in data_rows :
    if len(str(i)) == 1 :
        i_str = "00" + str(i)
    elif len(str(i)) == 2 :
        i_str = "0" + str(i)
    else :
        i_str = str(i)
    # end if
    #
    ezdoc = ezdxf.new('R2013', setup=True)
    ezmsp = ezdoc.modelspace()
    ezdoc.layers.new(name='Section-Plate', 
                     dxfattribs={'linetype': 'CONTINUOUS', 
                                 'lineweight': 150,
                                 'color': 1})
    #
    # text height default is 0.25
    text_str = i_str
    text_pos = (4.0, -1.1)
    ezmsp.add_lwpolyline(row, dxfattribs={'layer': 'Section-Plate'})
    ezmsp.add_text(text_str, 
                   dxfattribs={ 
                   'style': 'LiberationSerif',
                   'height': 0.5,
                   'color': 3}
                  ).set_pos(text_pos, align='MIDDLE_RIGHT')
    #
    file_id = i_str + '.dxf'
    ezdoc.saveas(path_graphic + "Laser-cut/SSNP-row-" + file_id)
    print('row is', i_str)
    i = i + 1
    ezdoc.layers.remove('Section-Plate')
# end for
#
print("Done")
```
