```python
# stl-SSNP-12B5-cut-plates-pickle-write-3.py
# 2021-12-28
#
# pickle 是 Python 的內建模組
# read 349x207x2_XZ.dat file  <-- derived from 349x196x3_XYZ.dat (X,Z) elements
#
import os
path = os.getcwd()
print('當前工作目錄 ==>', path)
#
from shapely.geometry  import LineString
from operator import itemgetter
#
import pprint
import pickle
#
path_data    = './data/'
#
def do_A(row,point2D_list) :
    #
    check_sw = "F" 
    for j, xy in enumerate(row) :
        if check_sw == "T" :
            point2D_list.append(row[j])              
        elif row[j+1][1] > 0.0 :
            check_sw = "T"
            point2D_list.append(row[j])   
        # end if
    # end for    
# end def
#
## ================================================================
#
with open(path_data + '349xvaryx2_XY.dat', 'rb') as in_file :
    in_rows = pickle.load(in_file)
# end with    
#
# trimed xy_rows
xy_rows = []
for i, row in enumerate(in_rows) :
    y_max = max(row,key=itemgetter(1))[1]
    if y_max > 0.0 :
        point2D_list = []
        [point2D_list.append(xy) for xy in row[:6]]
        do_A(row,point2D_list)
        xy_rows.append(point2D_list)
    else :
        xy_rows.append(row)
    # end if
    #
# end for
#
# gap is 0.5 cm
left_open_cut_xy  = [(10.25, -2) , (10.25, -1.2), ( 9.75, -1.2), ( 9.75, -2)]
left_dn_xy  = [(-1.0,-2.0)]
#
right_dn_xy = [(38.0,-2.0)]
right_open_cut_xy = [(28.25, -2) , (28.25, -1.2), (27.75, -1.2), (27.75, -2)]
#
new_xy_rows = []
for i in range(0,349,1) :
    Boundary = left_open_cut_xy + left_dn_xy + xy_rows[i] + right_dn_xy + right_open_cut_xy + [(10.25, -2)]
    new_xy_rows.append(Boundary)
# end for
#
print('Laser cut boundary has', len(xy_rows), 'rows')
print('         each rows has varying columns')
print ('\n')
#
# pprint.pprint(xy_rows[100])
# xy_rows --> shape=(349,varying,2)
with open(path_data + '349xvaryx2_XY.dat', 'wb') as out_file :
    pickle.dump(new_xy_rows, out_file)
# end with    
#
"""
print ('Done')
```
