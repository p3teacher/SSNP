```python
# stl-SSNP-14-ezDXF-mesh.py
# 2021-12-12
#
# pickle 是 Python 的內建模組
# read 349x196x3_XYZ.dat file
#
import os
path = os.getcwd()
print('當前工作目錄 ==>', path)
#
import pickle
import ezdxf
from ezdxf import units
#
path_data  = './data/'
path_3Dstl = './3Dstl/'
#
"""
# AC1027 ---> AutoCAD R2013
# line weight ---> 0.13mm = 13
# minimum line weight 13 
# maximum line width 200
#
# color  0 --> Black
# color  1 --> red
# color  3 --> green
# color  5 --> Dark blue
# color  6 --> Magenta
"""
#
## ================================================================
#
# up layer verteices
A_up = [38.0, 90.0, 0.0] # 0
B_up = [-1.0, 90.0, 0.0] # 1
C_up = [-1.0, 20.4, 0.0] # 2
D_up = [38.0, 20.4, 0.0] # 3
#
# down layer verteices
a_dn = [38.0, 90.0,-2.0] # 4
b_dn = [-1.0, 90.0,-2.0] # 5
c_dn = [-1.0, 20.4,-2.0] # 6
d_dn = [38.0, 20.4,-2.0] # 7
#
left_dn_xy  = (-1.0,-2.0)
left_up_xy  = (-1.0, 0.0)
# ... here is the dem boundary line
right_up_xy = (38.0, 0.0)
right_dn_xy = (38.0,-2.0)
#
left_open_cut_xy  = [(10.15, -2) , (10.15, -1.2), ( 9.85, -1.2), ( 9.85, -2)]
right_open_cut_xy = [(28.15, -2) , (28.15, -1.2), (27.85, -1.2), (27.85, -2)]
#
ezdoc = ezdxf.new('R2013', setup=True)
# Set centimeter as document/modelspace units
ezdoc.units = units.CM
# which is a shortcut (including validation) for
ezdoc.header['$INSUNITS'] = units.CM  # 5
#
ezmsp = ezdoc.modelspace()
mesh  = ezmsp.add_mesh()
#
with open(path_data + '349x196x3_XYZ.dat', 'rb') as in_file :
    dem_rows = pickle.load(in_file)
# end with    
#
xyz_rows = []
for i in range(0,349,1) :
    xyz_rows.append([tuple(xyz) for xyz in dem_rows[i]])
# end for
#
with mesh.edit_data() as mesh_tri:
    for i in range(0,348,1):
        for j in range(0,195,1):
            mesh_tri.add_face([xyz_rows[i][j],  xyz_rows[i]  [j+1], xyz_rows[i+1][j]])
            mesh_tri.add_face([xyz_rows[i][j+1],xyz_rows[i+1][j+1], xyz_rows[i+1][j]])
        # end for
        print('row', i)
    # end for
# end with
#
ezdoc.saveas(path_3Dstl + "SSNP_ezDXF_tri_mesh.dxf")
#
print("Done")
```
