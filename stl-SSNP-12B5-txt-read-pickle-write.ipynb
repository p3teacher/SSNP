```python
# stl-SSNP-12B5-txt-read-pickle-write.py
# 2022-01-22
#
# pickle 是 Python 的內建模組
# read 349-v2-2_XY.txt file (Hand adjusted file)
#
import os
path = os.getcwd()
print('當前工作目錄 ==>', path)
#
import csv
import pprint
import pickle
#
path_data    = './data/'
#
xy_rows = []
with open(path_data + '349-v2-2_XY.txt', 'r', encoding='UTF-8') as in_file :
    for line in in_file:
        xy_rows.append(eval(line.rstrip()))
    # end for
# end with
#
# check xy_rows
#
print('Laser cut boundary has', len(xy_rows), 'rows')
print('         each rows has varying columns')
print ('\n')
#
print(xy_rows[100])
print(xy_rows[100][:10])
# xy_rows --> shape=(349,varying,2)
#
with open(path_data + '349-v2-2_XY.dat', 'wb') as out_file :
    pickle.dump(xy_rows, out_file)
# end with    
#
print ('Done')
```
